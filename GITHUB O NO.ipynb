```python
# =====================================================
# BUSCADOR INTELIGENTE DE METAPELITAS AKFM
# =====================================================

import os
from PIL import Image
import matplotlib.pyplot as plt

# -------------------------------------------
# ALIAS
# -------------------------------------------

alias = {

    "biotita":"bt",
    "granate":"grt",
    "estaurolita":"st",
    "clorita":"chl",
    "cloritoide":"cld",
    "cianita":"ky",
    "sillimanita":"sil",
    "cordierita":"crd",
    "moscovita":"ms",
    "muscovita":"ms",
    "feldespato":"kfs",
    "feldespato_potasico":"kfs",
    "ortopiroxeno":"opx",

    "bt":"bt",
    "grt":"grt",
    "st":"st",
    "chl":"chl",
    "cld":"cld",
    "ky":"ky",
    "sil":"sil",
    "crd":"crd",
    "ms":"ms",
    "kfs":"kfs",
    "opx":"opx"
}

# -------------------------------------------
# PESOS PETROLÓGICOS
# -------------------------------------------

pesos = {

    "chl":1,
    "ms":1,

    "bt":2,

    "cld":4,
    "grt":4,

    "st":5,

    "ky":6,
    "sil":6,

    "crd":7,

    "kfs":8,

    "opx":10
}

# -------------------------------------------
# GRADO METAMÓRFICO RELATIVO
# SOLO PARA METAPELITAS
# -------------------------------------------

grado = {

    "A":"Muy bajo",
    "B":"Muy bajo",
    "C":"Muy bajo",
    "D":"Bajo",

    "E":"Bajo",
    "F":"Bajo",

    "G":"Medio",
    "H":"Medio",

    "I":"Medio",
    "J":"Medio",

    "K":"Medio-Alto",
    "L":"Medio-Alto",

    "M":"Alto",
    "N":"Alto",
    "O":"Alto",
    "P":"Alto",

    "Q":"Muy alto",
    "R":"Muy alto",

    "S":"Muy alto",
    "T":"Muy alto",

    "U":"Extremo"
}

# -------------------------------------------
# MINERALES AJENOS A METAPELITAS
# -------------------------------------------

minerales_maficos = {

    "act",
    "actinolita",
    "hbl",
    "hornblenda",
    "ep",
    "epidota",
    "pl",
    "plagioclasa",
    "di",
    "diopsido"
}

# -------------------------------------------
# INGRESO
# -------------------------------------------

print("\nMinerales observados:\n")

entrada = input(
    "Ingrese minerales separados por espacios:\n"
)

tokens = entrada.lower().split()

minerales_usuario = set()

for t in tokens:

    if t in alias:
        minerales_usuario.add(alias[t])

# -------------------------------------------
# VERIFICACIÓN
# -------------------------------------------

maficos = []

for t in tokens:

    if t in minerales_maficos:
        maficos.append(t)

if len(maficos) > 0:

    print("\n⚠ ADVERTENCIA")
    print("Existen minerales más típicos de metabasitas:")
    print(", ".join(maficos))

# -------------------------------------------
# ZONA ÍNDICE
# -------------------------------------------

zona = "Indeterminada"

if "opx" in minerales_usuario:
    zona = "Zona ortopiroxeno"

elif "crd" in minerales_usuario:
    zona = "Zona cordierita"

elif "sil" in minerales_usuario:
    zona = "Zona sillimanita"

elif "ky" in minerales_usuario:
    zona = "Zona cianita"

elif "st" in minerales_usuario:
    zona = "Zona estaurolita"

elif "grt" in minerales_usuario:
    zona = "Zona granate"

elif "bt" in minerales_usuario:
    zona = "Zona biotita"

elif "chl" in minerales_usuario:
    zona = "Zona clorita"

print("\n================================")
print("INTERPRETACIÓN")
print("================================")
print("Zona índice:", zona)

# -------------------------------------------
# BUSCAR TRIÁNGULOS
# -------------------------------------------

resultados = []

for nombre, minerales in triangulos.items():

    coincidencias = minerales_usuario.intersection(minerales)

    puntaje = sum(
        pesos.get(m,1)
        for m in coincidencias
    )

    faltantes = len(
        minerales_usuario - minerales
    )

    resultados.append(
        (
            nombre,
            puntaje,
            len(coincidencias),
            faltantes
        )
    )

# -------------------------------------------
# ORDENAR
# -------------------------------------------

resultados.sort(
    key=lambda x:(x[1],x[2],-x[3]),
    reverse=True
)

# -------------------------------------------
# MOSTRAR TOP 5
# -------------------------------------------

print("\n================================")
print("TOP 5 TRIÁNGULOS MÁS PROBABLES")
print("================================\n")

for r in resultados[:5]:

    nombre = r[0]
    puntaje = r[1]
    ncoinc = r[2]

    print(
        f"{nombre} | "
        f"Puntaje={puntaje} | "
        f"Coincidencias={ncoinc} | "
        f"Grado={grado[nombre]}"
    )

# -------------------------------------------
# MOSTRAR IMÁGENES
# -------------------------------------------

fig, axs = plt.subplots(1,5,figsize=(18,4))

for ax, r in zip(axs, resultados[:5]):

    nombre = r[0]

    ruta = f"triangulos_N2/{nombre}.png"

    if os.path.exists(ruta):

        img = Image.open(ruta)

        ax.imshow(img)

        ax.set_title(
            f"{nombre}\n{grado[nombre]}"
        )

    ax.axis("off")

plt.tight_layout()
plt.show()
```
